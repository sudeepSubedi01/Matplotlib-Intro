```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Subplots
- more flexible way to create subplots <br>
- allows to control figure size, share axes and return the individual axes to plot them directly
- Syntax <br>
  fig, axes = plt.subplots(nrows=2, ncols=2, figsize=)
- **fig** : Figure object ; it is the container holding all the subplots and other visual elements like title
- **axes** : Axes object ; represents the individual subplots where the actual plotting happens
- **ncols=2** : means figure will have 2 subplots arranged side by side
- **nrows=2** : means figure will have 2 subplots arranges up and down
- **figsize=(10,4)** : size of entire figure in inches (width x height)

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

fig, (ax1,ax2) = plt.subplots(ncols=2,figsize=(10,4))
# Plot 1
ax1.scatter(x,y)
ax1.set_title('Scatter')
ax1.set_xlabel('x-data')
ax1.set_ylabel('y-data')

# Plot 2
ax2.plot(x,y,color='red')
ax2.set_title('Line')
ax2.set_xlabel('x-data')
ax2.set_ylabel('y-data')

plt.tight_layout()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

fig, (ax1,ax2) = plt.subplots(nrows=2,figsize=(4,6))
# Plot 1
ax1.scatter(x,y)
ax1.set_title('Scatter')
ax1.set_xlabel('x-data')
ax1.set_ylabel('y-data')

# Plot 2
ax2.plot(x,y,color='red')
ax2.set_title('Line')
ax2.set_xlabel('x-data')
ax2.set_ylabel('y-data')

plt.tight_layout()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

fig, [[ax1, ax2],[ax3, ax4]] = plt.subplots(ncols=2,nrows=2, figsize=(10,10))
# (ax1,ax2,ax3,ax4) = axs.ravel()
# Plot 1
ax1.scatter(x,y)
ax1.set_title('Scatter')
ax1.set_xlabel('x-data')
ax1.set_ylabel('y-data')

# Plot 2
ax2.plot(x,y,color='red')
ax2.set_title('Line')
ax2.set_xlabel('x-data')
ax2.set_ylabel('y-data')

# Plot 3
ax3.scatter(x,y,color='green')
ax3.set_title('Scatter')
ax3.set_xlabel('x-data')
ax3.set_ylabel('y-data')

# Plot 4
ax4.plot(x,y,color='purple')
ax4.set_title('Line')
ax4.set_xlabel('x-data')
ax4.set_ylabel('y-data')

plt.tight_layout()
plt.show()
```

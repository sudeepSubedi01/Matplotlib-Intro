# Subplot
allows to display multiple plots in single figure arranged in grid like structure <br>
useful when we want to compare different datasets or vizualize multiple graphs side by side

```python
x = [1,2,3]
y = [1,3,8]
```

```python
plt.subplot(2,2,1) # figure has 2 row,2 col and this plot is 1st subplot
plt.scatter(x,y)
plt.title('Plot 1')

plt.subplot(2,2,2)
plt.plot(x,y)
plt.title('Plot 2')

plt.subplot(2,2,3)
plt.plot(x,y)
plt.title('Plot 3')

plt.tight_layout()    # ensures there is enough space between subplots to prevent overlap
plt.show()
```

# plt.figure() <br>
- creates a new figure 
- figure is a container that holds all elements like axes, labels, lines
- figure is like a blank canvas where we can add out plots,charts,vizualization

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6,4))

plt.plot([1,2,3,4],[10,20,30,20])
plt.show()
```

```python
plt.figure(num=1, figsize=(6,4), dpi=100)    # num is optional
plt.plot([1,2,3],[1,4,9])
plt.title('Figure 1')

plt.figure(num=2, figsize=(5,5))
plt.scatter([1,2,3],[1,4,9], color='red')
plt.title('Figure 2')
```
